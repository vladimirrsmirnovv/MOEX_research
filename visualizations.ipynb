### Визуализация 1

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Подготовка данных
data = {
    "Индекс": [
        "Финансов (MOEXFN)",
        "Нефти и газа (MOEXOG)",
        "Недвижимости (MOEXRE)",
        "Информационных технологий (MOEXIT)",
        "Металлов и добычи (MOEXMM)"
    ],
    "Объём торгов (₽)": [
        6406953283,
        5590327027,
        2404110528,
        1958714132,
        1468119203
    ]
}

df = pd.DataFrame(data)
df["Объём торгов (млрд ₽)"] = df["Объём торгов (₽)"] / 1e9

# Построение графика
fig, ax = plt.subplots(figsize=(10, 8))

sns.barplot(
    data=df,
    x="Индекс",
    y="Объём торгов (млрд ₽)",
    color='#c20430',
    edgecolor='black',
    linewidth=1.5,
    zorder=3,
    ax=ax
)

# Подписи над столбцами
for i, value in enumerate(df["Объём торгов (млрд ₽)"]):
    ax.text(
        i,
        value + 0.1,
        f"{value:.3f}",
        ha='center',
        va='bottom',
        fontsize=11,
        fontweight='bold',
        zorder=4
    )

# Настройка осей и графика
ax.set_xlabel('Индекс и код индекса', fontsize=12, color='black')
ax.set_ylabel('Объём торгов (млрд, ₽)', fontsize=12, color='black')
ax.set_title('Объёмы торгов по отраслевым индексам Московской биржи', fontsize=14, fontweight='bold')

# Только стилизация подписей оси X без set_xticklabels
ax.tick_params(axis='x', labelrotation=20, labelsize=10, colors='gray')
ax.tick_params(axis='y', labelsize=10)

# Границы графика
for spine in ['bottom', 'left']:
    ax.spines[spine].set_color('black')
    ax.spines[spine].set_linewidth(1.2)

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

# Сетка
ax.grid(axis='y', linestyle='--', alpha=0.5, zorder=1)
ax.set_axisbelow(True)

plt.tight_layout()
plt.show()
```

| Блок                          | Код  | Капитализация, RUB |
| ----------------------------- | ---- | ------------------ |
| **Финансов**                  | SBER | 6 593 949 136 080  |

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Данные для трёх компаний
data = {
    "Сектор": [
        "Финансов (SBER)",
        "Нефти и газа (GAZP)",
        "Строительства (PIKK)",
    ],
    "Тикер": [
        "SBER",
        "GAZP",
        "PIKK"
    ],
    "Капитализация (₽)": [
        6593949136080,      # SBER
        3261263137104,      # GAZP
        319482565292.8       # PIKK
    ]
}
df = pd.DataFrame(data)
df["Капитализация (трлн ₽)"] = df["Капитализация (₽)"] / 1e12

# Сортировка по убыванию капитализации
df = df.sort_values("Капитализация (трлн ₽)", ascending=False).reset_index(drop=True)

# Построение графика
fig, ax = plt.subplots(figsize=(8, 6))

sns.barplot(
    data=df,
    x="Тикер",
    y="Капитализация (трлн ₽)",
    color='#00833f',
    edgecolor='black',
    linewidth=1.5,
    zorder=3,
    ax=ax
)

# Устанавливаем правильные метки
ax.set_xticks(range(len(df)))
ax.set_xticklabels(df["Сектор"], rotation=20, fontsize=10, color='gray')

# Подписи над столбцами
for i, value in enumerate(df["Капитализация (трлн ₽)"]):
    ax.text(
        i,
        value + 0.05,
        f"{value:.3f}",
        ha='center',
        va='bottom',
        fontsize=11,
        fontweight='bold',
        zorder=4
    )

# Настройка осей и графика
ax.set_xlabel('Сектор и тикер компании', fontsize=12, color='black')
ax.set_ylabel('Капитализация (трлн ₽)', fontsize=12, color='black')
ax.set_title('Капитализация крупнейших компаний по секторам', fontsize=14, fontweight='bold')

ax.tick_params(axis='y', labelsize=10)

# Границы графика
for spine in ['bottom', 'left']:
    ax.spines[spine].set_color('black')
    ax.spines[spine].set_linewidth(1.2)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

# Сетка
ax.grid(axis='y', linestyle='--', alpha=0.5, zorder=1)
ax.set_axisbelow(True)

plt.tight_layout()
plt.show()
```
